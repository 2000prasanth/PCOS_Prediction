```python
from pingouin import canonical_correlation_analysis

# Assuming df is your DataFrame and 'binary_column' is the binary column
binary_column = 'PCOS (Y/N)'

# Create two DataFrames for each class
df_class_0 = df[df[binary_column] == 0]
df_class_1 = df[df[binary_column] == 1]

# Drop the binary column
df_class_0 = df_class_0.drop(columns=[binary_column])
df_class_1 = df_class_1.drop(columns=[binary_column])

# Compute canonical correlation
cca_results = canonical_correlation_analysis(df_class_0, df_class_1)

# Print canonical correlation coefficients
print("Canonical Correlation Coefficients:")
print(cca_results['r'])

# Print canonical coefficients
print("\nCanonical Coefficients for Set 1:")
print(cca_results['coef_X'])

print("\nCanonical Coefficients for Set 2:")
print(cca_results['coef_Y'])
```

```python
import pandas as pd
from statsmodels.multivariate.cancorr import CanCorr
import numpy as np

# Assuming df is your DataFrame and 'binary_column' is the binary column
#binary_column = 'binary_column'

# Create two DataFrames for each class
df_class_0 = df[df[binary_column] == 0]
df_class_1 = df[df[binary_column] == 1]

# Drop the binary column
df_class_0 = df_class_0.drop(columns=[binary_column])
df_class_1 = df_class_1.drop(columns=[binary_column])

# Convert non-numeric values to NaN
df_class_0 = df_class_0.apply(pd.to_numeric, errors='coerce')
df_class_1 = df_class_1.apply(pd.to_numeric, errors='coerce')

# Remove rows with NaN values
df_class_0.dropna(inplace=True)
df_class_1.dropna(inplace=True)

# Perform canonical correlation analysis
cca = CanCorr(df_class_0.values, df_class_1.values)
cca_results = cca.cancorr()

# Print canonical correlation coefficients
print("Canonical Correlation Coefficients:")
print(cca_results)

# Print canonical coefficients
print("\nCanonical Coefficients for Set 1:")
print(cca.x_cancorr)

print("\nCanonical Coefficients for Set 2:")
print(cca.y_cancorr)
```
